````python
import re


def has_mbti(text):
    mbtis = '[infj|entp|intp|intj|entj|enfj|infp|enfp|isfp|istp|isfj|istj|estp|esfp|estj|esfj]+'

    if re.findall(mbtis, text):
        return True
    else:
        return False
    
    
df.content.apply(has_mbti)
````

````python
#å°†ç»“æœå­˜å‚¨åˆ°dfä¸­
df['hasMBTI'] = df['content'].apply(has_mbti)
df
````

è¿™æ ·åç»­çš„åˆ†æä»»åŠ¡ï¼Œå°±å¯ä»¥é€šè¿‡æŸ¥çœ‹mbtiå­—çœ¼å‰åå‡ºç°çš„å­—ç¬¦ï¼Œæ¥æ›´æ–°æ­£åˆ™è¡¨è¾¾å¼ã€‚

**æ­£åˆ™è¡¨è¾¾å¼å«ä¹‰**

```
mbti_win = "(.{0,5}(?:infj|entp|intp|intj|entj|enfj|infp|enfp|isfp|istp|isfj|istj|estp|esfp|estj|esfj).{0,5})"
```
- ``(`` å’Œ ``)``è¿™äº›æ‹¬å·ç”¨äºå°†æ•´ä¸ªåŒ¹é…ç»“æœæ•è·ä¸ºä¸€ä¸ªåˆ†ç»„
- ``.{0,5}`` ï¼šè¿™æ˜¯ä¸€ä¸ªé‡è¯ï¼Œè¡¨ç¤ºåŒ¹é…å‰é¢çš„å­—ç¬¦ï¼ˆ.è¡¨ç¤ºåŒ¹é…ä»»æ„å­—ç¬¦ï¼‰é›¶æ¬¡åˆ°äº”æ¬¡ã€‚è¿™éƒ¨åˆ†ç”¨äºåŒ¹é…å‰é¢çš„æ–‡æœ¬ï¼Œç¡®ä¿æœ€å¤šåŒ¹é…å‰é¢çš„äº”ä¸ªå­—ç¬¦ã€‚
- ``(?:infj|entp|intp|intj|entj|enfj|infp|enfp|isfp|istp|isfj|istj|estp|esfp|estj|esfj)``ï¼šè¿™æ˜¯ä¸€ä¸ªéæ•è·åˆ†ç»„ï¼Œç”¨äºå°†å¤šä¸ªMBTIç±»å‹è¯æ±‡ç”¨ | è¿æ¥èµ·æ¥ï¼Œè¡¨ç¤ºåŒ¹é…å…¶ä¸­ä»»ä½•ä¸€ä¸ªã€‚
- ``.{0,5}`` ï¼šè¿™éƒ¨åˆ†åŒæ ·æ˜¯ä¸€ä¸ªé‡è¯ï¼Œè¡¨ç¤ºåŒ¹é…åé¢çš„å­—ç¬¦ï¼Œç¡®ä¿æœ€å¤šåŒ¹é…åé¢çš„äº”ä¸ªå­—ç¬¦ã€‚

<br>

````python
def mbti_window(text):
    #è¯†åˆ«mbtiçš„æ­£åˆ™è¡¨è¾¾å¼ 
    mbti_win = "(.{0,5}(?:infj|entp|intp|intj|entj|enfj|infp|enfp|isfp|istp|isfj|istj|estp|esfp|estj|esfj).{0,5})"

    try:
        return re.findall(mbti_win, text)[0]
    except:
        return "æœªè¯†åˆ«"

df['MBTI_win'] = df['content'].apply(mbti_window)
df
````

å¯ä»¥åŸºäºæ­¤è®¾è®¡ä¸€ä¸ªä¸¥æ ¼çš„æ­£åˆ™è¡¨è¾¾å¼ï¼Œèƒ½è¯†åˆ«åˆ°çš„è®°å½•ï¼Œè‚¯å®šèƒ½åˆ¤æ–­è¯¥ç”¨æˆ·çš„mbtiç±»å‹ã€‚ æœªè¯†åˆ«åˆ°çš„æ ‡è®°ä¸º "æœªè¯†åˆ«"

```python
mbti_regex = "[æˆ‘|è‡ªå·±|å˜æˆ|ä¸€ç›´|æ˜¯|æœ¬]*(infj|entp|intp|intj|entj|enfj|infp|enfp|isfp|istp|isfj|istj|estp|esfp|estj|esfj)[æˆ‘|ä¿º|æœ¬|è‡ªå·±]*"
```

- ``[æˆ‘|è‡ªå·±|å˜æˆ|ä¸€ç›´|æ˜¯|æœ¬]*``ï¼šè¿™éƒ¨åˆ†æ˜¯ä¸€ä¸ªå­—ç¬¦é›†åˆï¼Œç”¨äºåŒ¹é…å‰é¢çš„å­—ç¬¦ï¼ˆå…³é”®è¯ï¼‰ã€‚æ–¹æ‹¬å· ``[...]`` è¡¨ç¤ºå­—ç¬¦ç±»ï¼Œå…¶ä¸­çš„å­—ç¬¦æ˜¯å¯é€‰çš„ï¼Œå¹¶ä¸” * è¡¨ç¤ºåŒ¹é…é›¶æ¬¡æˆ–å¤šæ¬¡ã€‚è¿™æ„å‘³ç€å®ƒå¯ä»¥åŒ¹é…é›¶ä¸ªæˆ–å¤šä¸ªå‡ºç°åœ¨æ–¹æ‹¬å·ä¸­çš„å­—ç¬¦ï¼Œä¾‹å¦‚å¯ä»¥åŒ¹é…"æˆ‘"ã€"è‡ªå·±"ã€"å˜æˆ"ã€"ä¸€ç›´"ã€"æ˜¯"ã€"æœ¬"ç­‰è¿™äº›å…³é”®è¯ã€‚
- ``(infj|entp|intp|intj|entj|enfj|infp|enfp|isfp|istp|isfj|istj|estp|esfp|estj|esfj)`` ï¼šè¿™æ˜¯ä¸€ä¸ªåˆ†ç»„ï¼Œå…¶ä¸­åŒ…å«äº†MBTIç±»å‹è¯æ±‡ï¼Œç”¨ç«–çº¿ ``|`` åˆ†éš”ï¼Œè¡¨ç¤º"æˆ–"çš„å…³ç³»ã€‚è¿™éƒ¨åˆ†ç”¨äºåŒ¹é…ä»»æ„ä¸€ä¸ªMBTIç±»å‹è¯æ±‡ã€‚
- ``[æˆ‘|ä¿º|æœ¬|è‡ªå·±]*`` ï¼šè¿™éƒ¨åˆ†ä¸ç¬¬1éƒ¨åˆ†ç±»ä¼¼ï¼Œæ˜¯ä¸€ä¸ªå­—ç¬¦é›†åˆï¼Œç”¨äºåŒ¹é…åé¢çš„å­—ç¬¦ï¼ˆå…³é”®è¯ï¼‰ã€‚åŒæ ·ï¼Œæ–¹æ‹¬å· ``[...]`` è¡¨ç¤ºå­—ç¬¦ç±»ï¼Œå…¶ä¸­çš„å­—ç¬¦æ˜¯å¯é€‰çš„ï¼Œå¹¶ä¸” ``*`` è¡¨ç¤ºåŒ¹é…é›¶æ¬¡æˆ–å¤šæ¬¡ã€‚

````python

def identify_mbti(text):
    if '//@' in text:
        new_text = text.split('//@')[0]
    else:
        new_text = text

    #è¯†åˆ«mbtiçš„æ­£åˆ™è¡¨è¾¾å¼ 
    mbti_regex = "[æˆ‘|è‡ªå·±|å˜æˆ|ä¸€ç›´|æ˜¯|æœ¬]*(infj|entp|intp|intj|entj|enfj|infp|enfp|isfp|istp|isfj|istj|estp|esfp|estj|esfj)[æˆ‘|ä¿º|æœ¬|è‡ªå·±]*"

    try:
        return re.findall(mbti_regex, text)[0]
    except:
        return "æœªè¯†åˆ«"

#mbtiç±»å‹
df['MBTI_Cat'] = df['content'].apply(identify_mbti)
df
````
